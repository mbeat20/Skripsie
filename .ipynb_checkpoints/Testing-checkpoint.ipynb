```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Set start and duration for audio loading
start = 1  # start at 1 second
duration = 2 * 60  # duration of 2 minutes

# Load the audio file
audio_data, sampling_rate = librosa.load('data/ElephantIsland2014/wav/20140906_110000_AWI251-01_AU0231_250Hz.wav', sr=100, offset=start, duration=duration)

# Apply pre-emphasis (optional)
audio_data = librosa.effects.preemphasis(audio_data)

# Compute Mel Spectrogram
mel_spectrogram = librosa.feature.melspectrogram(y=audio_data, sr=sampling_rate, n_mels=128, fmax=4000)

# Convert Mel Spectrogram to dB scale
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Plot the Mel Spectrogram
plt.figure(figsize=(15, 5))
librosa.display.specshow(mel_spectrogram_db, sr=sampling_rate, x_axis='time', y_axis='mel', cmap='viridis')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.ylim(0,30)
plt.show()

# Compute MFCCs from the Mel spectrogram
num_mfcc = 12
mfccs = librosa.feature.mfcc(S=mel_spectrogram, sr=sampling_rate, n_mfcc=num_mfcc)

# Standardize the MFCC features
scaler = StandardScaler()
mfccs_scaled = scaler.fit_transform(mfccs.T)

# Plot the MFCCs
plt.figure(figsize=(10, 6))
librosa.display.specshow(mfccs, sr=sampling_rate, x_axis='time')
plt.colorbar(label='MFCC Coefficients')
plt.title(f'MFCC Spectrogram with {num_mfcc} Coefficients')
plt.xlabel('Time (s)')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()

plt.show()

# Apply K-means clustering
k = 2  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=0).fit(mfccs_scaled)
labels = kmeans.labels_

# Plot the MFCCs with cluster labels
plt.figure(figsize=(15, 5))
librosa.display.specshow(mfccs_scaled.T, sr=sampling_rate, x_axis='time')
plt.scatter(range(mfccs.shape[1]), labels, c=labels, cmap='viridis', s=50)
plt.colorbar(label='Cluster Label')
plt.title('MFCCs with Cluster Labels')
plt.xlabel('Time (s)')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()
plt.show()
```
